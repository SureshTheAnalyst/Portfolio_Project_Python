# Data analysis using Covid-19 dataset with python. 
## The data contains daily confirmed cases and daily death cases for 100 days from January 22, 2020 to April 30, 2020.
------------------------------------------

### 1. First, let's import the necessary modules 

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

### 2. Importing covid-19 dataset

```python
corona_dataset_csv = pd.read_csv("covid19_confirmed_dataset.csv")
corona_dataset_csv.head(10)
```

#### Let's check the shape of the dataframe

```python
corona_dataset_csv.shape
```

### 2. Delete the non-essential columns

```python
df = corona_dataset_csv.drop(["Lat", "Long", "Province/State"], axis = 1)
```

```python
df.head(10)
```

### 3. Aggregating the rows by the country

```python
corona_dataset_aggregated = df.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

#### Checking the shape again

```python
corona_dataset_aggregated.shape
```

### 4. Visualizing data related to a country for example China
visualization always helps for better understanding of our data.

```python
corona_dataset_aggregated.dtypes
```

```python
corona_dataset_aggregated.loc["China"].plot()
```

```python
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["Canada"].plot()
plt.legend()
```

### 5. Calculating a good measure 
we need to find a good measure reperestend as a number, describing the spread of the virus in a country. 

```python
corona_dataset_aggregated.loc["China"][:3].plot()
```

### 6. Caculating and plotting the first derivative of the curve

```python
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
corona_dataset_aggregated.loc["China"].diff().plot()
corona_dataset_aggregated.loc["India"].diff().plot()
corona_dataset_aggregated.loc["US"].diff().plot()
plt.legend()
```

### 7. Finding maxmimum infection rate for China

```python
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
corona_dataset_aggregated.loc["US"].diff().max()
```

### 8. Finding maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
max_infection_rates
```

### 9. Inserting this into a new column

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["max_infection_rate"] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

### 10. Creating a new dataframe with only needed column

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
```

```python
corona_data.head()
```

### Next, we are 
- Importing the WorldHappinessReport.csv dataset
- selecting needed columns for our analysis 
- joining the datasets 
- calculating the correlations as the result of our analysis

### 12. Importing the WorldHappinessReport.csv dataset

```python
happiness_report_csv = pd.read_csv("worldwide_happiness_report.csv")
```

```python
happiness_report_csv.head()
```

### 13. Let's drop the unnecessary columns

```python
drop_cols = ["Overall rank", "Score", "Generosity", "Perceptions of corruption"]
```

```python
happiness_report = happiness_report_csv.drop(drop_cols, axis=1)
happiness_report.head()
```

### 14. Changing the indices of the dataframe

```python
happiness_report = happiness_report.set_index("Country or region")
```

```python
happiness_report.head()
```

### 15. Let's join two dataset we have prepared

```python
data = corona_data.join(happiness_report, how="inner")
data.head()
```

```python
data.shape
```

### 16. Combine another table

```python
combined = corona_death_data.join(data, how="inner")
combined.head()
```

### 17. correlation matrix 

```python
combined.corr()
```
